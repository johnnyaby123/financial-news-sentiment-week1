```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import talib
```

```python
data_path = '../data/'
```

```python
stock_files = ['AAPL.csv', 'AMZN.csv', 'GOOG.csv', 'META.csv', 'MSFT.csv', 'NVDA.csv']
stock_list = []

for file in stock_files:
    df = pd.read_csv(os.path.join(data_path, file))
    df['Date'] = pd.to_datetime(df['Date'])
    df['ticker'] = file.replace('.csv','')  # Add ticker column
    stock_list.append(df)

stock_df = pd.concat(stock_list, ignore_index=True)
stock_df = stock_df.sort_values(['ticker','Date'])
```

```python
stock_df['return'] = stock_df.groupby('ticker')['Close'].pct_change()
stock_df.dropna(inplace=True)
```

```python
indicators_df = pd.DataFrame()

for ticker in stock_df['ticker'].unique():
    df = stock_df[stock_df['ticker'] == ticker].copy()
    
    # Simple Moving Average
    df['SMA_10'] = talib.SMA(df['Close'], timeperiod=10)
    
    # Exponential Moving Average
    df['EMA_10'] = talib.EMA(df['Close'], timeperiod=10)
    
    # Relative Strength Index
    df['RSI_14'] = talib.RSI(df['Close'], timeperiod=14)
    
    # MACD
    df['MACD'], df['MACD_signal'], df['MACD_hist'] = talib.MACD(df['Close'])
    
    indicators_df = pd.concat([indicators_df, df], ignore_index=True)
```

```python
indicators_df.head()
```

```python
for ticker in indicators_df['ticker'].unique():
    df = indicators_df[indicators_df['ticker']==ticker]
    
    plt.figure(figsize=(12,5))
    plt.plot(df['Date'], df['Close'], label='Close')
    plt.plot(df['Date'], df['SMA_10'], label='SMA 10')
    plt.plot(df['Date'], df['EMA_10'], label='EMA 10')
    plt.title(f"{ticker} Close Price and Moving Averages")
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
    
```

```python
processed_path = os.path.join(data_path, 'stock_indicators.csv')
indicators_df.to_csv(processed_path, index=False)
print(f"Processed data saved to {processed_path}")
```
