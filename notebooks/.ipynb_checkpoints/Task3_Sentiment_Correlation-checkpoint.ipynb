```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
%matplotlib inline
```

```python
data_path = '../data/'
```

```python
stock_df = pd.read_csv(data_path + 'stock_indicators.csv')
stock_df['Date'] = pd.to_datetime(stock_df['Date'])
```

```python
stock_df['return'] = stock_df.groupby('ticker')['Close'].pct_change()
stock_df.dropna(inplace=True)
```

```python
stock_df['daily_sentiment'] = 0.0
```

```python
daily_sentiment_df = stock_df[['Date','ticker','daily_sentiment']].copy()
```

```python
merged_df = pd.merge(stock_df, daily_sentiment_df,
                     on=['Date','ticker'],
                     how='left')
```

```python
if 'daily_sentiment' not in stock_df.columns:
    stock_df['daily_sentiment'] = 0.0  
```

```python
print(stock_df.head())
```

```python
print("Correlation of daily returns vs news sentiment per stock (placeholder values):")
for ticker in stock_df['ticker'].unique():
    df = stock_df[stock_df['ticker'] == ticker]
    corr = df['return'].corr(df['daily_sentiment'])
    print(f"{ticker}: {corr:.4f}")
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=stock_df, x='daily_sentiment', y='return', hue='ticker')
plt.title('Daily Stock Returns vs News Sentiment (placeholder)')
plt.xlabel('Daily News Sentiment')
plt.ylabel('Daily Stock Return')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.show()
```

```python
stock_df.to_csv(data_path + 'task3_workflow_ready.csv', index=False)
print(f"Workflow-ready data saved to {data_path + 'task3_workflow_ready.csv'}")
```
